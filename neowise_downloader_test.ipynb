```python
import pyvo
import time
# count = 0


def download():
    service = pyvo.dal.TAPService('https://irsa.ipac.caltech.edu/TAP')
    start = time.time()
    result = service.run_async("""
       SELECT ra,dec,mjd,w1mpro,w1sigmpro,w1snr,nb,na,w1sat,cc_flags,saa_sep,moon_masked
       FROM neowiser_p1bs_psd
       WHERE (w1mpro<12 and sso_flg==0 qual_frame>0 and qi_fact>0.9 and saa_sep>0) 
    """)
    tab = result.to_table()
    tab.write('./test_data/neowise_w1_brighter_than_12.fits',format='fits')
    end = time.time()
    print("download finished,time use: %0.2f seconds"%(end-start))
```

```python
download()
```
