```python
from util import make_single_light_curve
import numpy as np
import pickle

def sigma_finding(target_list,index):
    i = 0
    while i < len(target_list):
        long_name = target_list.iloc[i]['long_name']
        id_in_matched = target_list.iloc[i]['id_in_matched']
        name = long_name[0:3]
        table1 = pd.read_csv('./mached_catalog/'+name+'/'+long_name+'/'+long_name+'_w1_mached.csv')
        
        ra,dec,mag,error,mjdmean = make_single_light_curve(table1,int(id_in_matched))

        sigma = np.std(mag)
        ex_1_mag = np.array([elmt for j,elmt in enumerate(mag) if j != np.argmax(mag)])
        ex_1_sigma = np.std(ex_1_mag)
        ex_2_mag = np.array([elmt for j,elmt in enumerate(ex_1_mag) if j != np.argmin(ex_1_mag)])
        ex_2_sigma = np.std(ex_2_mag)

        print(1,sigma,ex_1_sigma,ex_2_sigma)
        i+=1
```
