```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

def simbad_type_count(filename):
    table =  pd.read_csv(filename)
    print(len(table))
    from collections import Counter
    main_types = table['main_type'].tolist()
    counter = Counter(main_types)
    print(counter)
    total_number = np.sum(list(counter.values()))
    percentages = list(counter.values()) / total_number * 100
    percebtages_counter = {}
    keys = sorted(counter, key=counter.get, reverse=True)
    percentages = sorted(percentages, reverse=True)
    for i,key in enumerate(keys):
        percebtages_counter[key] = percentages[i]

    for i in range(len(percebtages_counter)):
        print(f"{list(percebtages_counter.keys())[i]}: {list(percebtages_counter.values())[i]:.2f}%")
```

```python
simbad_type_count('./results/simbad_above_0.7_new_w1.csv')
```
