```python
import glob
import os
import pandas as pd
first_layer_names = glob.glob('[0-9][0-9][0-9]',root_dir='./mached_catalog')
first_layer_names.sort()
count=0
percentage=0
for name in first_layer_names:
    second_layer_names = glob.glob('[0-9][0-9][0-9][0-9]*',
                                    root_dir='./mached_catalog/'+name)
    second_layer_names.sort()
    for long_name in second_layer_names:
        band='w1'
        if True == os.path.isfile('./mached_catalog/'+name+'/'+long_name+'/'+long_name+'_'+ band +'_new_features.csv'):  
            table = pd.read_csv('./mached_catalog/'+name+'/'+long_name+'/'+long_name+'_'+ band +'_new_features.csv')
            length = len(table.columns) 
            percentage+=1
            print(percentage*100/18000,'/%')          
            if length != 18:
                print(length, long_name)
                count+=1
print(count)
```
