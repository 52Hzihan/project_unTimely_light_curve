```python
import pandas as pd
# table = pd.read_csv('total_catalogues/w1_var_para_table1.csv')
proto_tables = []
for i in range(0,9):
    proto_tables.append(pd.read_csv('total_catalogues/w1_var_para_table%d.csv'%i))
```

```python
# relibility_table = table[table['median']<16.5]
# len(relibility_table)
```

```python
variable_candidate_table_list = []
for table in proto_tables:
    variable_candidate_table_list.append(table[table['Eta_e']>7])
```

```python
reliable_candidate_list = []
for table in variable_candidate_table_list:
    reliable_candidate_list.append(table[table['median']<16.5])
```

```python
print(len(reliable_candidate_list))
```

```python
reliable_candidate_table = pd.concat(reliable_candidate_list)
print(len(reliable_candidate_table))
```

```python
reliable_candidate_table.to_csv('./total_catalogues/Eta_e_above_7_max_mag16_5.csv')
```

```python
light_sources = variable_candidate_table[variable_candidate_table['median']<15.5]
print(len(light_sources))
```
